```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import t

sns.set()
sns.set_style("dark")
sns.set_color_codes()

n = 24
alpha = .05
ddof = n - 1

# right tail
t_crit = t.ppf(q=1 - alpha, df=23)

# test statistic value
t_val = 2.24


# plot t-distribution
x = np.arange(-4,4,.01)
y = t.pdf(x, ddof)
plt.figure(figsize=(8, 4))
plt.title('t-Distribution\n%s degrees of freedom' % ddof)
plt.plot(x, y)

# show t-critical value
plt.axvline(x=t_crit)
plt.text(t_crit, t.pdf(t_crit, ddof), 'crit val=%.4f, alpha=%.2f' % (t_crit, alpha))

# shade area above critical value
x_tc = np.arange(t_crit, 4, .01)
y_tc = t.pdf(x_tc, ddof)
plt.fill_between(x_tc, y_tc, alpha=.3)

# plot test statistic
plt.axvline(x=t_val)
plt.text(t_val, t.pdf(t_val, ddof), 'test stat=%.4f' % t_val)
```
